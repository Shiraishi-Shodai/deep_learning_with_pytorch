# 3.5 テンソル要素の型

```python
import torch
```

## 3.5.3 テンソルのdtype属性の管理

```python
double_points = torch.ones(10, 2, dtype=torch.double)
short_points = torch.ones(10, 2, dtype=torch.short)
print(double_points)
print(short_points)
```

### ２種類のキャスト
**to関数**  
変換可能か確認してから変換する

```python
# ➀
double_points = torch.ones(10, 2).double()
short_points = torch.ones(10, 2).short()
print(double_points)
print(short_points)

# ➁
double_points = torch.ones(10, 2).to(torch.double)
short_points = torch.ones(10, 2).to(torch.short)
print(double_points)
print(short_points)
```

### 演算で入力型が混同するときは、入力は自動的に大きい型に変換される

```python
points_64 = torch.rand(5, dtype=torch.double)
points_short = points_64.to(torch.short)
points_64 * points_short
```
