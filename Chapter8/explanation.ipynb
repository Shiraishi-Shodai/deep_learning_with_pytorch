# 第８章 畳み込み

## 全結合層  
層の入力となる各ニューロンが、次の層のすべてのニューロンに接続されている。モデルの最終段階で使用されることが多い
* メリット  
  * 複数の特徴を組み合わせることが得意  
* デメリット  
  * 過学習になりやすい
  * パラメータの数が多い
  * 空間的な位置情報を持たない
  * 汎化性が低い  

## 畳み込み
カーネルを画像全体にスライドさせ、各ピクセルと重みの加重和を計算することで画像の特徴を抽出する。  
カーネルを使って抽出した画像の局所的な特徴を組み合わせて画像全体の特徴を抽出

* メリット
  * 局所的な演算処理
  * 移動不変性
  * パラメータを大幅に削減

## メモ
* 移動不変性  
オブジェクトの位置が変わっても、同じように認識できる能力を指します。例えば、画像内の猫が画像の左端にいても右端にいても、システムがそれを「猫」と認識できる能力です。
* 局所性  
入力データの小さな局所的な領域（例えば、画像の一部分）に注目すること
* カーネル  
特定の特徴に反応する小さなウィンドウ
